<font color='blue'>
    <font size = 4>
    In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. 
    The E-Commerce company has the following rule for computing their own revenue: </font> 
    
#####     The company charges  </font> 
#####  (i) 25% on the orders having final price (discounted price) greater than 600 </font> 

##### (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600 </font> 

##### (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350 </font> 

##### (iv) Otherwise, 5% on the final price (discounted price)    </font> 

<font color = 'blue'>
    <font size = 4>
        Net revenue generated by the E-Commerce company over all orders placed.
    </font>

```python
ecom_revenue=[]
for i in Ecom_Data['discounted_price']:    
    if i>600:                              
        c = (i*(25/100))                   
        ecom_revenue.append(c)
    elif i>350 and i<=600:                  
        c = (i*(15/100)) 
        ecom_revenue.append(c)            
    elif i>100 and i<=350:                 
        c = (i*(10/100))   
        ecom_revenue.append(c)            
    else:                                 
        c = (i*(5/100))    
        ecom_revenue.append(c)             
print ("The Net revenue generated by E-commerce company is :", round(sum(ecom_revenue),3))
```

*Net BrandRevenue is BrandRevenue which is generated after all the deductions. Also, BrandRevenue is different from the E-Commerce company's net revenue.*

```python
Ecom_Data['ecom_revenue']=ecom_revenue
Ecom_Data['brand_revenue']=Ecom_Data['discounted_price']- Ecom_Data['ecom_revenue']
```

```python
#listing the top 14 NetBrandRevenue each Brand
Ecom_Data.groupby(['Brand'])['brand_revenue'].sum().reset_index().sort_values(by='brand_revenue',ascending=False).head(14)
```
